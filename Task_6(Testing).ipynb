## Hypothesis Testing

#### Hypothesis is an assumption of population parameter.

#### in testing basically we are testing the claim either we accept it or reject it.

### there are two types of hypothesis :-
### 1) Null Hypothesis: 
#### a) Null Hypothesis means no change we take claim as it is.
#### b) We denotes null hypothesis as H0.
### 2) Alternative Hypothesis :-
#### a) Alternative hypothesis means opposite of null hypothesis.
#### b) We denotes alternative hypothesis as H1.
#### c) it is ay be two tail or one tail.

## Errors in testing

### Type l error 
#### type I error is the mistaken rejection of an actually true null hypothesis.

### Type - ll error
#### type lI error is the mistaken acceptance of an actually false null hypothesis.

## Level of significance

#### Popular levels of significance are 10% (0.1), 5% (0.05), 1% (0.01), 0.5% (0.005), and 0.1% (0.001). If a test of significance gives a p-value lower than or equal to the significance level, the null hypothesis is rejected at that level.

## P-value

#### 1) the p-value is the probability of obtaining test results at least as extreme as the results actually observed.
#### 2) smaller p-value means that there is stronger evidence in favor of the alternative hypothesis.
#### 3) If a test of significance gives a p-value lower than or equal to the significance level, the null hypothesis is rejected at that level.¶
