# Biomial distribution


```python
# Scratch code for binomial distribuion
# we need to define 1st factorial
def fact(n):
    x=1
    for i in range(1,n+1):
        x=x*i
    return x
# then we need to define combination
def combi(n,x):
    a=fact(n)
    b=fact(x)
    c=fact(n-x)
    combi=a/(b*c)
    return combi
# now we will write code for bnomial distribution
def binomial(n,p,x):
    return combi(n,x)*(p)**(x)*(1-p)**(n-x)
```

### Example :

```python
binomial(20,0.5,10)
```

# Bernoulli distribution


```python
def bernoulli(x,p):
    ber = p**x*(1-p)**(1-x)
    return ber
```

# Exponential distribution


```python
def expo(x,lamb):
    return lam * np.exp(-lam * x)
```

# Uniform distribution
in this distribution every value have an equal chance to occure in given range.


```python
def uniform(x , a , b):
    if a <= x <= b:
        return 1/(b-a)
    else :
        return 0
```

### Example :

```python
uniform(0.5,0,1)
```

# Normal distrubution 


```python
def normal(x,mu,sigma):
    return (1 / (sigma*np.sqrt(2*np.pi)))*np.exp((-1/2)*((x-mu)/sigma)**2)
```

# Poisson Distribution
The formula for Poisson distribution is 


```python
def poisson(x, lamb):
    return ((lamb**x)*(np.exp**-lamb))/fact(x)
```
